## Exerc√≠cio 7

```python
#filtra o dataset pelo artista dolly-parton
songs_most_present_singer = data[data['artist'] == most_present_singer]
```

```python
list_words_most = []

for i in songs_most_present_singer['lyrics']:
    #transforma a letra atual em uma lista de palavras
    letra_atual = i.lower().replace('\n',' ' ).split(" ")
    
    contador = {}
    for j in letra_atual:
        if j in contador:
            contador[j] += 1
        else:
            contador[j] = 1
            
    #transforma dicionario em uma lista ordenada pelo valor
    words_ordenado = sorted(contador.items() ,key=lambda t: t[1] ,reverse =True)
    #pega as 5 primeiras 
    first_five = [i[0] for i in words_ordenado[0:5]]
    
    #transforma a lista em uma string
    string_words = ""
    for i in first_five:
        string_words += (i + ",")
    
    list_words_most.append(string_words)

songs_most_present_singer.insert(6, "most_used_word", list_words_most)
songs_most_present_singer.head()
```

```python
words_data =  pd.read_csv('mood_words.csv', sep=',', encoding='latin-1')
words_data.head()
```

```python
clasficador = []

#isso aqui ta muito feio n√£o e pra ser assim
negativo = list(words_data['negative_words'])
positivo = list(words_data['positive_words'])

for i in songs_most_present_singer['lyrics']:
    #transforma a letra atual em uma lista de palavras
    letra_atual = i.lower().replace('\n',' ' ).split(" ")

    #conta as palavras da atual
    contador = {'negativo': 0, 'positivo':0}
    
    for j in letra_atual:
        if j in negativo:
            contador['negativo'] += 1
        elif j in positivo:
            contador['positivo'] += 1
        else:
            pass
    
    if contador['positivo'] > contador['negativo']:
        clasficador.append("positivo")
    elif contador['positivo'] < contador['negativo']:
        clasficador.append("negativo")
    else:
        clasficador.append("neutro") #quando √© igual

        
songs_most_present_singer = songs_most_present_singer.drop('mood', axis=1) # n√£o tem que tirar essa linha?
songs_most_present_singer.insert(7, "mood", clasficador)
songs_most_present_singer.head()
```

```python
## Pega quantidade de m√∫sicas positivas e negativas do cantor mais presente.

## Cria playlists de musicas positivas e negativas

mood_songs = Counter(songs_most_present_singer["mood"])

mood_songs_positive = songs_most_present_singer.loc[songs_most_present_singer['mood'] == "positivo"] 
mood_songs_negative = songs_most_present_singer.loc[songs_most_present_singer['mood'] == "negativo"] 

positive_mood = len(mood_songs_positive)
negative_mood = len(mood_songs_negative)
neutral_mood = mood_songs["neutro"]
```

```python
## Plota gr√°fico pizza de m√∫sicas do artista
import matplotlib.pyplot as plt

labels = 'Positive', 'Negative', 'Neutral'
sizes = [positive_mood, negative_mood, neutral_mood]
colors = ['yellowgreen', 'lightblue', 'gold']
 
# Plot
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%')
 
plt.axis('equal')
plt.title('Moods of ' + most_present_singer + "'s songs")
plt.show()
```
